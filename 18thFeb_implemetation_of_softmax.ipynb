```
%matplotlib inline
import matplotlib.pyplot as plt
import torch
import torchvision
import helper
import numpy as np
```

```

from torchvision import datasets, transforms

transform = transforms.Compose([transforms.ToTensor(),
                               transforms.Normalize((0.5,0.5,0.5),(0.5,0.5,0.5)),
                                ])

trainset = datasets.MNIST('MNIST_data/', download=True, train=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=64, shuffle=True)

```

```
def activation(x):
  return 1/(1+torch.exp(-x))
```

```
def NeuralNetwork(images, labels):
  inputs = images.view(images.shape[0], -1)
  
  inputUnits = inputs.shape[1]
  hiddenUnits = 256
  outputUnits = 10
  
  w1 = torch.randn((inputUnits, hiddenUnits))
  w2 = torch.randn((hiddenUnits, outputUnits))
  
  b1 = torch.randn(hiddenUnits)
  b2 = torch.randn(outputUnits)
  
  # Forward propogate from input layer to hidden layer
  
  x1 = activation(torch.mm(inputs,w1)+b1)
  x2 = activation(torch.mm(x1,w2)+b2)
  
  out = torch.sum(torch.exp(x2),dim=1).view(-1,1)
  e = torch.exp(x2)
  finalOut = e / out
  print(out.shape, e.shape, finalOut.shape)
  
```

```
data = iter(trainloader)
images, labels = data.next()
NeuralNetwork(images,labels)
```
