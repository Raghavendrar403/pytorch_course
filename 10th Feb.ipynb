```python
def sigmoid(x):
    return (1/(1+np.exp(-x)))
```

```python
class NeuralNetwork:
    
    def __init__(self, inputNodes, hiddenNodes, outputNodes, learningRate):
        # Inititalizing the no.of nodes in input, hidden and output layer
        self.inputNodes = inputNodes
        self.outputNodes = outputNodes
        self.hiddenNodes = hiddenNodes
        
        # Initializing the learning rate
        self.learningRate = learningRate
        
        # Initialize the weights to the Neural Network
        self.wih = (np.random.rand(self.hiddenNodes, self.inputNodes) - 0.5)
        self.who = (np.random.rand(self.outputNodes, self.hiddenNodes) - 0.5)
        pass
    
    
    def train(self, inputList, outputList):
        # Convert both input and output lists to numpy arrays with 2 dimensions
        inputs = np.array(inputList, ndmin=2).T
        outputs = np.array(outputList, ndmin=2).T
       
        # The below 2 steps are called forward propogation
        # Step 1 - Calculate the values of nodes at the hidden layer
        hiddenInputs = np.dot(self.wih, inputs)
        hiddenOutputs = sigmoid(hiddenInputs)
        
        # Step 2 - Calculate the values of nodes at the final layer
        finalInputs = np.dot(self.who, hiddenOutputs)
        finalOutputs = sigmoid(finalInputs)
                                      
        # Step 3 - Calculate error (target-actual)
        error = outputs - finalOutputs
        hidden_error = np.dot(self.who.T, error)
        
        # Step 4 - Update Weights of the output layer and hidden layer
        self.who += self.learningRate * np.dot(error * finalOutputs * (1.0-finalOutputs), hiddenOutputs.T)
        self.wih += self.learningRate * np.dot(hidden_error * hiddenOutputs * (1-hiddenOutputs), inputs.T)
        pass
    

    
    def querry(self, inputs):
        # Calculate the values of nodes at the hidden layer
        hiddenInputs = np.dot(self.wih, inputs)
        hiddenOutputs = sigmoid(hiddenInputs)
        
        # Calculate the values of nodes at the final layer
        finalInputs = np.dot(self.who, hiddenOutputs)
        finalOutputs = sigmoid(finalInputs)
        return finalOutputs
        pass

    
```
