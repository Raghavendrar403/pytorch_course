```
import torch
import torchvision
import numpy
```

```
class Network(torch.nn.Module):
  def __init__(self):
    super().__init__()
    
    # Create hidden and output layer
    self.hidden1 = torch.nn.Linear(784,128)
    self.hidden2 = torch.nn.Linear(128,64)
    self.output = torch.nn.Linear(64,10)
    
    # Create a sigmoid and softmax
    self.relu = torch.nn.ReLU()
    self.softmax = torch.nn.Softmax(dim=1)
    
  def forward(self,x):
    x = self.hidden1(x)
    x = self.relu(x)
    x = self.hidden2(x)
    x = self.relu(x)
    x = self.output(x)
    x = self.softmax(x)
    
    return x
```
